### Motivation

Here in this notebook, we will implement Expectation Maximization Algorithm with Gaussian Mixture Model. We will take $3$ clusters with each cluster having $100$ data points.

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
from matplotlib import colors
from matplotlib import cm
%matplotlib inline
```

```python
num_cluster = 3
num_sample_per_cluster = 100
num_total_sample = num_cluster * num_sample_per_cluster
```

Lets create some gaussian samples and for reproductibility we are using random seed. Here we have used wishart random variable to create initial covariance matrix.

```python
np.random.seed(8000)

samples = []

for _ in range(num_cluster):
    
    mean = np.random.normal(size = 2) * 10
    
    cov = st.wishart.rvs(3, scale = [[1,0], [0,1]])
    
    samples.append(np.random.multivariate_normal(mean, cov, size = num_sample_per_cluster))
    
samples = np.concatenate(samples, axis = 0)

print(samples.shape)
```

Plot the samples

```python
plt.figure(figsize = [10,10])
plt.scatter(samples[:, 0], samples[:, 1])
plt.show()
```

Expectation Maximization:

E-step

$\gamma\left(z_{n k}\right)=\frac{\pi_{k} N\left(x_{n} | \mu_{k}, \Sigma_{k}\right)}{\sum_{j=1}^{K} \pi_{j} N\left(x_{n} | \mu_{j}, \Sigma_{j}\right)}$

```python
def e_step(pi, mean, cov, sample):
    
    r_ic = np.zeros([num_total_sample, num_cluster])
    
    for k in range(num_cluster):
        
        r = pi[k] * st.multivariate_normal.pdf(sample, mean = mean[k], cov = cov[k] + 1e-5 * np.eye(2))
        
        r_ic[:, k] = r
        
    r_ic = r_ic / np.sum(r_ic, axis = 1)[:, None]
    
    return r_ic
```

M-step

$\pi_{k}^{n e w}=\frac{N_{k}}{N} \\
\mu_{k}^{n e w}=\frac{1}{N_{k}}\sum_{n=1}^{N} \gamma\left(z_{n k}\right) x_{n} \\
\Sigma_{k}^{n e w}=\frac{1}{N_{k}} \sum_{n=1}^{N} \gamma\left(z_{n k}\right)\left(x_{n}-\mu_{k}^{n e w}\right)\left(x_{n}-\mu_{k}^{n e w}\right)^{T}$

```python
def m_step(r_ic, sample):
    
    for k in range(num_cluster):
        
        n_k = np.sum(r_ic[:, k])
        
        pis[k] = n_k / num_total_sample
        
        means[k] = np.sum(r_ic[:, k][:, None] * sample, axis =0) / n_k
        
        diffs = (sample - means[k])[:, :, None]
        
        cov = np.zeros([2, 2])
        
        for n in range(num_total_sample):
            cov += r_ic[n, k] * diffs[n] @ diffs[n].T
            
        covs[k] = cov / n_k
        
    return pis, means, covs
```

Plotting Function:

Reference:

1. [Matplotlib](https://matplotlib.org)

```python
def confidence_ellipse(ax, pi, mean, cov, n_std=3.0, facecolor='none', **kwargs):
    
    pearson = cov[0, 1]/np.sqrt(cov[0, 0] * cov[1, 1])

    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0),
        width=ell_radius_x * 2,
        height=ell_radius_y * 2,
        facecolor=facecolor,
        alpha=pi * 0.5,
        **kwargs)

    scale_x = np.sqrt(cov[0, 0]) * n_std

    scale_y = np.sqrt(cov[1, 1]) * n_std

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean[0], mean[1])

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)

def plot(ax, samples, r_ic, pis, means, covs, title):
    cluster_colors = cm.get_cmap('Set1').colors
     
    for k in range(num_cluster):
        rgba = colors.to_rgba(cluster_colors[k % len(cluster_colors)])
        rgba_n = np.array([rgba for _ in range(num_total_sample)])
        rgba_n[:, 3] = r_ic[:, k]
        
        ax.scatter(samples[:, 0], samples[:, 1], color=rgba_n)
        ax.scatter(means[k, 0], means[k, 1], marker='x', color='black', alpha=min(pis[k] * num_cluster, 1.0))
        confidence_ellipse(ax, pis[k], means[k], covs[k], facecolor=rgba, edgecolor=rgba)
    
    ax.set_title(title)
```

```python
pis = np.ones(num_cluster) / num_cluster

means = np.random.normal(size=(num_cluster, 2))

covs = [np.array([[1, 0], [0, 1]]) for _ in range(num_cluster)]

for i in range(50):
    
    r_ic = e_step(pis, means, covs, samples)
    
    if i%10 == 0:
        print('iteration: {}'.format(i))
        fig, ax = plt.subplots(1, 2, figsize=[10, 4])
        fig.suptitle('iteration : ' +str(i))
        plot(ax[0], samples, r_ic, pis, means, covs, 'E-Step')
        
    pis, means, covs = m_step(r_ic, samples)
    
    if i%10 == 0:
        
        plot(ax[1], samples, r_ic, pis, means, covs, 'M-Step')
        plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 8])
plot(ax, samples, r_ic, pis, means, covs, 'Final')
plt.show()
```
