### Basic Gaussian Processing

The purpose of this notebook is to implement a basic gaussian distribution and to show how the kernel functions can smooth the distributions. Here we have implemented squared exponential kernel.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
sb.set()
%matplotlib inline
```

```python
n = 50

x = np.linspace(0,10,n).reshape(-1,1)
```

$x \sim N(sin(x), I)$

```python
norm = st.multivariate_normal(np.sin(x).ravel(), np.eye(n)*1e-6)
norm
```

```python
plt.figure(figsize=(16,6))
plt.plot(x, norm.rvs())
```

Multivariate Normal Sampling using $mean = 0$ and $variance = I$



$x \sim N(0, I)$

```python
sample = st.multivariate_normal(np.zeros(n).ravel(), np.eye(n))
```

```python
plt.figure(figsize=(16,6))
for _ in range(3):
    plt.plot(x, sample.rvs())
plt.title('3 Samples')
```

### Kernels

Squared exponential Kernel

$\kappa(\mathbf{x}_i,\mathbf{x}_j) = \sigma_f^2 \exp(-\frac{1}{2l^2}
  (\mathbf{x}_i - \mathbf{x}_j)^T
  (\mathbf{x}_i - \mathbf{x}_j))$

```python
def kern(x1, x2, l=1.0, sigma_f=1.0):

    exp_part = np.sum(x1**2, 1).reshape(-1, 1) + np.sum(x2**2, 1) - 2 * np.dot(x1, x2.T)
    return sigma_f**2 * np.exp(-0.5 / l**2 * exp_part)
```

```python
cov = kern(x, x, .44)
sample = st.multivariate_normal(np.zeros(n).ravel(), cov)
```

```python
plt.figure(figsize=(16,6))
for _ in range(3):
    plt.plot(x, sample.rvs())
plt.title(r'sampled from $N(0,k(x,x))$')
```
