## Motivation

We are going to implement Markov Chain Monte Carlo from scratch and will try implement the Matropolis-Hasting algorithm. Matropolis-Hasting works good with higher-dimensional spaces. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
from copy import copy

sb.set()
%matplotlib inline
```

```python
# creating a method for form a population with mean 10 and std 3

def mod1(t):
    return np.random.normal(10, 3, t)
```

```python
np.random.seed(123)
population = mod1(30000)

# randomly taking 10000 samples

observations = population[np.random.randint(0, 30000, 10000)]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.hist(observations, bins= 35,)
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Observations of 10000 samples')
mu_obs = observations.mean()
print(mu_obs)
```

Functions like: prior, transition model, acceptance, metropolis hasting are here.

Here $x[0] = \mu$ and $x[1] = \sigma$

for log_like_normal function here is the equation:

$\sum_i^n -nLog(\sigma_{current}\sqrt{2\pi})-\dfrac{(d_i-\mu_{obs})^2}{2\sigma_{current}^2}$

```python
def trans_model(x):
    
    return [x[0], np.random.normal(x[1], 0.5, (1, ))[0]]

def prior(x):

    if(x[1] <=0):
        return 0
    
    return 1


def log_like_normal(x,data):

    return np.sum(-np.log(x[1] * np.sqrt(2* np.pi) )-((data-x[0])**2) / (2*x[1]**2))


def acceptance(x, x_new):
    
    if x_new>x:
        return True
    else:
        accept=np.random.uniform(0,1)
        return (accept < (np.exp(x_new-x)))


def metropolis_hastings(likelihood_computation, prior, transition_model, param_init, iterations, data, acceptance_rule):
    
    x = param_init
    accepted = []
    rejected = []   
    for i in range(iterations):
        x_new =  transition_model(x)    
        x_lik = likelihood_computation(x, data)
        x_new_lik = likelihood_computation(x_new, data) 
        if (acceptance_rule(x_lik + np.log(prior(x)), x_new_lik + np.log(prior(x_new)))):            
            x = x_new
            accepted.append(x_new)
        else:
            rejected.append(x_new)            
                
    return np.array(accepted), np.array(rejected)
```

```python
accepted, rejected = metropolis_hastings(log_like_normal, prior, trans_model, [mu_obs, 0.1], \
                                         10000, observations, acceptance)
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(2,1,1)

ax.plot(rejected[0:100,1], 'rx', label='Rejected',alpha=0.5)
ax.plot(accepted[0:100,1], 'b.', label='Accepted',alpha=0.5)
ax.set_xlabel("Iteration")
ax.set_title("First 100 samples are shown.")
ax.grid()
ax.legend()



ax2 = fig.add_subplot(2,1,2)
to_show=-accepted.shape[0]
ax2.plot( rejected[to_show:,1], 'rx', label='Rejected',alpha=0.5)
ax2.plot( accepted[to_show:,1], 'b.', label='Accepted',alpha=0.5)
ax2.set_xlabel("Iteration")
ax2.set_title("All samples are shown.")
ax2.grid()
ax2.legend()

fig.tight_layout()
```
