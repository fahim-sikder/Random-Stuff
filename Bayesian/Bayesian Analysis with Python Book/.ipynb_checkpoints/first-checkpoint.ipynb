```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
sb.set()
%matplotlib inline
```

```python
mu_param = [-1, 0, 1]
sd_param = [0.5, 1, 1.5]

x = np.linspace(-7, 7, 200)

_, ax = plt.subplots(3, 3, figsize = (15,7))

for i in range(3):
    for j in range(3):
        mu = mu_param[i]
        sd = sd_param[j]
        
        y = st.norm(mu,sd).pdf(x)
        
        ax[i,j].plot(x, y)
        ax[i,j].plot([], label = r'$\mu$ = {:3.2f} $\sigma$ = {:3.2f}'.format(mu,sd), alpha = 0)
        ax[i,j].legend(loc = 1)
```
